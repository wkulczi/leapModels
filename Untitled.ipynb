```python
def show_dataset_samples(data_dir="imgs", filePaths=[],cmap='viridis'):
    plt.figure(figsize=(10,10))
    for i in range(len(filePaths)):
        plt.subplot(3,2,i+1)
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        img = mpimg.imread(f"{data_dir}/{filePaths[i]}")
        plt.imshow(img, cmap="gray")
        plt.xlabel(filePaths[i])
    plt.show()
```
